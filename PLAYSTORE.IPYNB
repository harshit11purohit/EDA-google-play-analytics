```python
def size_process(item):
    if str(item)[-1] == "M":
        res = float(str(item).replace("M", ""))
        res = res * 1024
        return res
    
    elif str(item)[-1] == "K":
        res = float(str(item).replace("K", ""))
        return res
    else: 
        return np.nan   # âœ… minimal fix
```

```python
size_process("45M")
```

```python
df_copy["Size"] = df_copy["Size"].apply(size_process)

df_copy["Size"].dtype 
#df_copy.info()
#@df_copy["Size"]
```
